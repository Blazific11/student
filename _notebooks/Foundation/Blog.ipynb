- Beginning Of The Year
At the beginning of the year, I’ll be honest — I wasn’t very good at computer science. Everything felt confusing, from writing simple code to setting up my projects. One of the first things I did was create a virtual environment, which at the time felt like a huge step. I didn’t totally understand what it did, but it helped me get started and gave me a place to experiment safely without breaking anything else on my computer. Also my focus wasnt the best.

- Middle of the Year
As the year went on, I started learning more and even helped teach small classes about JavaScript. Teaching others turned out to be one of the best ways to learn myself. I got to see how other people thought about coding problems, and I learned a lot from my classmates too. My focus at this time of the year had gotten better but I was still very easily distracted.

- Changing Games and Learning how CSS
One of the biggest turning points for me was when we started modifying games. Changing little parts of the games helped me understand how CSS worked, especially with visual effects like borders and animations. I remember discovering commands like borderGlow, which made elements on the screen light up in cool ways. That was when coding really started to feel creative and fun, not just technical. As more assignments started piling up my focus started improving and I wasnt as easily distracted anymore.

- End of the Year
By the end of the year, I had made my final game project. It was the most challenging and rewarding thing I’d done all year. I learned how to work with different game engines and how to use basic pixel art to make my designs come alive. Looking back, I’m proud of how far I’ve come — from struggling to set up a virtual environment to creating my own interactive game. This class taught me not just about coding, but also about learning, teamwork, and creativity.
